# preparing the twitter data

```python
import warnings
warnings.filterwarnings("ignore")
def get_twitter(city):
  way = '/content/drive/MyDrive/lab3ml/'
  way_feat = f'{city}Features/features{city}/Twitter/'
  lda50_features = pd.read_csv(way + way_feat + 'LDA50Features.csv')
  lda50_topic_terms = pd.read_csv(way + way_feat + 'descriptions/LDA50FeaturesTopicTerms.csv', encoding='unicode_escape')
  liwc_features = pd.read_csv(way + way_feat + 'LIWCFeatures.csv')
  manually_defined_features = pd.read_csv(way + way_feat + 'manuallyDefinedTextFeatures.csv')
  df_topics = lda50_features
  df_terms = lda50_topic_terms
  terms_dict = df_terms.set_index(['Topic id', 'Term'])['Weight'].to_dict()

  result = pd.DataFrame(df_topics['_id'])

  for topic_id in range(50):
      topic_column = f'topic_{topic_id}'
      if topic_column in df_topics.columns:
          topic_values = df_topics[topic_column]
          terms_for_topic = df_terms[df_terms['Topic id'] == topic_column]


          for term, weight in zip(terms_for_topic['Term'], terms_for_topic['Weight']):
              result[term] = topic_values * weight

  combined_data = pd.merge(manually_defined_features, liwc_features, on='_id', how='outer')
  combined_data = pd.merge(combined_data, result, on='_id', how='outer')
  for column in combined_data.columns:
      if combined_data[column].dtype in ['float64', 'int64']:
          mean_value = combined_data[column].mean()
          combined_data[column].fillna(mean_value, inplace=True)
      else:
          combined_data[column].fillna(combined_data[column].mode()[0], inplace=True)
  return combined_data
```

```python
def get_foursquare(city):
  way = '/content/drive/MyDrive/lab3ml/'
  way_feat = f'{city}Features/features{city}/Foursquare/'
  venue1Test = pd.read_csv(way + way_feat+'venueCategoriesFeatures1MonthTest.csv')
  venue2Test = pd.read_csv(way + way_feat+'venueCategoriesFeatures2MonthsTest.csv')
  venue2Train = pd.read_csv(way + way_feat+'venueCategoriesFeatures2MonthsTrain.csv')
  venue3 = pd.read_csv(way + way_feat+'venueCategoriesFeatures3MonthsTrain.csv')
  venue5 = pd.read_csv(way + way_feat+'venueCategoriesFeatures5Months.csv')
  venueLDA6 = pd.read_csv(way + way_feat+'venueCategoriesLDA6Features.csv')
  combined_data = pd.merge(venue5, venueLDA6, on='_id', how='outer')
  for column in combined_data.columns:
      combined_data[column].fillna(0, inplace=True)
  return combined_data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.stem import SnowballStemmer
import numpy as np

def add_info(data, word):
  stemmer = SnowballStemmer("english")

  def stem_words(words, stemmer):
      return [stemmer.stem(word) for word in words]
  vectorizer = TfidfVectorizer(analyzer='word', tokenizer=lambda text: stem_words(text.split(), stemmer))

  non_empty_education_info = data['educationInfoForAgeGroupEstimation'].dropna()


  tfidf_matrix = vectorizer.fit_transform(non_empty_education_info)


  feature_names = np.array(vectorizer.get_feature_names_out())


  school_related_idxs = np.where(np.array([stemmer.stem(word) in feature for feature in feature_names]))[0]

  contains_school = [any(tfidf_matrix[i, idx] > 0 for idx in school_related_idxs) for i in range(tfidf_matrix.shape[0])]

  result_contains_school = []
  for index, row in data.iterrows():
      if pd.isna(row['educationInfoForAgeGroupEstimation']):
          result_contains_school.append(False)
      else:
        result_contains_school.append(contains_school.pop(0))
  data[f'contains{word}'] = result_contains_school
```

# creating a model

```python
from catboost import CatBoostClassifier
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score


def test(X_train, y_train, X_test, y_test):
    model = CatBoostClassifier(n_estimators=10, max_depth=5, feature_border_type='GreedyLogSum', logging_level="Silent", cat_features=list(X_train.dtypes.reset_index()[X_train.dtypes.reset_index()[0] == 'object']['index']))
    model.fit(X_train, y_train, verbose=1)
    predict = model.predict(X_test)
    return f1_score(y_test, predict, average='weighted'), accuracy_score(y_test, predict)

def do_testing(funcs, task_feature):
  cityes_train = pd.DataFrame()
  cityes_test = pd.DataFrame()
  way = '/content/drive/MyDrive/lab3ml/'
  X = pd.DataFrame()
  for city in list(["London", "Singapore", "NY"]):
    city_train =  pd.read_csv(way+city+'Train.csv')
    city_test = pd.read_csv(way+city+'Test.csv')
    city_train.drop(city_train[task_feature].isna()[city_train[task_feature].isna()].index, inplace=True)
    city_test.drop(city_test[task_feature].isna()[city_test[task_feature].isna()].index, inplace=True)

    cityes_train = pd.concat([cityes_train, city_train], ignore_index=True)
    cityes_test = pd.concat([cityes_test, city_test], ignore_index=True)
    if city == "NY":
      city = "NewYork"

    combined_data = funcs[0](city).copy()
    if len(funcs)>1:
      for func in funcs[1:]:
        combined_data = pd.merge(combined_data, func(city), on='_id', how='inner')
    X = pd.concat([X, combined_data], ignore_index=True)
  train = pd.merge(left=X, right=cityes_train[[task_feature, 'row ID']], how='inner', left_on='_id', right_on='row ID').drop('row ID', axis=1)
  test_ = pd.merge(left=X, right=cityes_test[[task_feature, 'row ID']], how='inner', left_on='_id', right_on='row ID').drop('row ID', axis=1)
  return test(train.drop(task_feature, axis=1), train[task_feature].fillna(cityes_train[task_feature]), test_.drop(task_feature, axis=1), test_[task_feature].fillna(cityes_test[task_feature]))
```

```python
def get_stats():
  tasks_feature = list(['relationship',	'educationLevel',	'gender'])
  for i in range(2):
    for j in range(2):
      for l in range(2):
        tasks = []
        variants = ''
        if i == 1:
          tasks.append(get_foursquare)
          variants+='foursquare '
        if j == 1:
          tasks.append(get_inst)
          variants+='instagram '
        if l == 1:
          tasks.append(get_twitter)
          variants+='twitter'
        if len(tasks)>0:
          variant_data = pd.DataFrame()
          count = 0
          for task in tasks_feature:
            count+=1
            result = do_testing(tasks, task)
            data = {"variant": variants, "task": task, "F1 Measure": result[0], "Accuracy": result[1]}
            variant_data = pd.concat([variant_data, pd.DataFrame(data=data, index=[count])])
          print(variant_data)
```
